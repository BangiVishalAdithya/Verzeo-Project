```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/arib168/data/main/50_Startups.csv")
df
```

```python
df.info()
```

```python
startups_numeric=df.select_dtypes(include=['int64','float64'])
startups_numeric
```

```python
x=startups_numeric.iloc[:,0:3].values
y=startups_numeric.iloc[:,-1].values
```

```python
x
```

```python
#TRAIN TEST AND SPLIT ON THE DATA TEST
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
model.predict([[116983.80,45173.06,14681.06]])
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from IPython.core.pylabtools import figsize
import matplotlib.pyplot as plt

plt.scatter(y_test,y_pred)
plt.xlabel('Actual')
plt.ylabel('predicted')
plt.title('Actual vs. predicted')
```
